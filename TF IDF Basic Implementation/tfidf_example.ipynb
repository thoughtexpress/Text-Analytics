```python
#specify the statements which will be considered as a document. in short, from now, Document == Statement
docA = "My name is Mango Man"
docB = "Mango Man currently resides in India"
docC = "Mango Man likes the season of summer"
docD = "Summers are amazing in India"
docE = "Mango is the fruit of the season summer"
docF = "Among the fruit, Mango fruit is called the king of fruit"
docG = "Apart from the Mango fruit out of all the fruits, watermelon fruit is the other fruit available in summer"
```

```python
#Making the Bag of Words (bow) for each document
#split function splits the document by the space
bowA = docA.split(" ")
bowB = docB.split(" ")
bowC = docC.split(" ")
bowD = docD.split(" ")
bowE = docE.split(" ")
bowF = docF.split(" ")
bowG = docG.split(" ")
```

```python
#bowE
```

```python
wordset = set(bowA).union(set(bowB)).union(set(bowC)).union(set(bowD)).union(set(bowE)).union(set(bowF)).union(set(bowG))
```

```python
type(wordset)
```

```python
wordDictA = dict.fromkeys(wordset, 0)
wordDictB = dict.fromkeys(wordset, 0)
wordDictC= dict.fromkeys(wordset, 0)
wordDictD = dict.fromkeys(wordset, 0)
wordDictE = dict.fromkeys(wordset, 0)
wordDictF = dict.fromkeys(wordset, 0)
wordDictG = dict.fromkeys(wordset, 0)
```

```python
#wordDictA
```

```python
for word in bowA:
    wordDictA[word]+=1
for word in bowB:
    wordDictB[word]+=1
for word in bowC:
    wordDictC[word]+=1
for word in bowD:
    wordDictD[word]+=1
for word in bowE:
    wordDictE[word]+=1
for word in bowF:
    wordDictF[word]+=1
for word in bowG:
    wordDictG[word]+=1
```

```python
#wordDictA
```

```python
import pandas as pd
pd.set_option('display.max_columns', 40)
pd.DataFrame([wordDictA,wordDictB,wordDictC,wordDictD,wordDictE,wordDictF,wordDictG])
```

```python
#computing the term frequency; which means the number of times the word
#is repeating in the document divided by the total number of words

#Formula: TF(word)= (number of word occuring in document/ total number of document)

def computeTF(wordDict, bow):
    tfDict = {}
    bowCount = len(bow)
    for word, count in wordDict.items():
        tfDict[word] = count/float(bowCount)
    return tfDict
```

```python
tfBowA = computeTF(wordDictA,bowA)
tfBowB = computeTF(wordDictB,bowB)
tfBowC = computeTF(wordDictC,bowC)
tfBowD = computeTF(wordDictD,bowD)
tfBowE = computeTF(wordDictE,bowE)
tfBowF = computeTF(wordDictF,bowF)
tfBowG = computeTF(wordDictG,bowG)
```

```python
#inverse document frequency
#this means how infrequent the word is
# higher the occurence of word, lower the IDF of that word will be

#Formula is IDF(word) = log(Number of documents/Occurence of word in separate Documents)

def computeIDF(docList):
    import math
    idfDict = {}
    N = len(docList)
    
    idfDict = dict.fromkeys(docList[0].keys(), 0)
    for doc in docList:
        for word, val in doc.items():
            if val > 0:
                idfDict[word] += 1
    
    for word, val in idfDict.items():
        idfDict[word] = math.log10(N / float(val))
        
    return idfDict
```

```python
idfs = computeIDF([wordDictA, wordDictB, wordDictC, wordDictD, wordDictE, wordDictF, wordDictG])
```

```python
def computeTFIDF(tfBow, idfs):
    tfidf = {}
    for word, val in tfBow.items():
        tfidf[word] = val*idfs[word]
    return tfidf
```

```python
tfidfBowA = computeTFIDF(tfBowA, idfs)
tfidfBowB = computeTFIDF(tfBowB, idfs)
tfidfBowC = computeTFIDF(tfBowC, idfs)
tfidfBowD = computeTFIDF(tfBowD, idfs)
tfidfBowE = computeTFIDF(tfBowE, idfs)
tfidfBowF = computeTFIDF(tfBowF, idfs)
tfidfBowG = computeTFIDF(tfBowG, idfs)
```
